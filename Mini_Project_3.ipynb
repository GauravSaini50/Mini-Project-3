Subset Selection Problem

1. Problem Statement:
Find all the subsets from a set of numbers whose sum is zero.

Constraint: Subset size must be 5
Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
import random as r

Set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
SetSize = 5
Iterations = 1000
ResultList = set()

for _ in range(Iterations):
    Chromosome = r.sample(Set, SetSize)
    if sum(Chromosome) == 0:
        # Convert to sorted tuple to avoid different orderings being considered unique
        # (Optional: If you want order to matter, just use tuple(Chromosome))
        ResultList.add(tuple(sorted(Chromosome)))

# Display results
print(f"Found {len(ResultList)} unique zero-sum subsets:")
for i, subset in enumerate(ResultList, 1):
    print(f"Subset {i}: {subset}")
```

2. Problem Statement:
Find all the subsets from a set of numbers whose sum is zero.

Constraint: Subset size must be 3 to 6 only
Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
import random as r

Set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
ResultList = {size: set() for size in range(3, 7)}  # For sizes 3 to 6
Iterations = 10000  # Number of random samples per size

for size in range(3, 7):
    for _ in range(Iterations):
        subset = tuple(sorted(r.sample(Set, size)))
        if sum(subset) == 0:
            ResultList[size].add(subset)

# Display results
for size in sorted(ResultList.keys()):
    print(f"\nSubsets of size {size} with sum 0:")
    for i, subset in enumerate(ResultList[size], 1):
        print(f"  Subset {i}: {subset}")
```
